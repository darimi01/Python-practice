```python
#for, while 등의 반복문 사용

a=int(input())

for i in range(1,10):
  print("{} * {} = {}".format(a,i,a*i))
```

```python
T=int(input())

for i in range(T):
  a,b=map(int,input().split())
  print("{}".format(a+b))
```

```python
n=int(input())
a=0 #0이라고 정의해줘야함. 안그러면 a에 쓰레기 값이 저장되서 숫자가 이상해짐

for i in range(1,n+1):#n+1을 해줘야 n까지 다 더해짐
  a = a+i
print(a)
```

```python
X=int(input())
N=int(input())
c=0

for i in range(N):
  a,b=map(int,input().split())
  c=(a*b)+c

if c==X:
  print("Yes")
else:
  print("No")
```

```python
N=int(input())

for i in range(1,N+1):
  print("*"*i)
```

```python
N=int(input())

for i in range(1,N+1):
  word='*'*i
  result = word.rjust(N)
  print(result)
#문자열을 오른쪽으로 정렬하려면 str.rjust() 메서드사용.
#왼쪽으로 정렬은 str.ljust()
```
