```python
#어렵다 이 문제
N,M=map(int,input().split())
baskets=[0]*N
for a in range(M):
  i,j,k=map(int,input().split())
  for b in range(i-1,j):
    baskets[b]=k #위에서 baskets을 0으로 지정해뒀기 때문에 k를 0으로 지정안해도되는듯

print(*baskets)#*은 리스트의 모든 요소를 개별적으로 출력하는데 사용

```

```python
N,M=map(int,input().split())
balls = list(range(1,N+1))

for _ in range(M):
  i,j=map(int,input().split())
  balls[i-1],balls[j-1]=balls[j-1],balls[i-1] #등가성립으로 예를들면 1=3,3=1을 통해 서로 교환하는것

print(*balls)
```

```python
N,M=map(int,input().split())
baskets=list(range(1,N+1))

for _ in range(M):
  i,j=map(int,input().split())
  baskets[i-1:j]=reversed(baskets[i-1:j]) #역순으로 바꾸는 메소드인 reversed

print(*baskets)
```

```python
scores=[]
f_scores=[]
#최댓값구하기
N=int(input())
for _ in range(N):
  a=int(input())
  scores.append(a)
b=max(scores)
#조작된 점수 만들기
for i in scores:
  c=(i/b)*100
  f_scores.append(c)
#평균구하기
d=sum(f_scores)/N

print(d)


#이건 챗gpt가 짠 코드
N = int(input())
scores = list(map(int, input().split()))

# 최댓값 구하기
max_score = max(scores)

# 점수를 새로 계산하여 리스트에 저장
new_scores = [(score / max_score) * 100 for score in scores]

# 평균 구하기
average = sum(new_scores) / N

print(average)
```
